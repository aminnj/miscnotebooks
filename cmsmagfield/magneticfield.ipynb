```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import os
from tqdm import tqdm

import pandas as pd
```

```python
dfc = pd.read_pickle("bfield_10cm_5deg.pkl.bz2")
dfc["x"] = dfc["rho"]*np.cos(dfc["phi"])
dfc["y"] = dfc["rho"]*np.sin(dfc["phi"])
dfc["bmag"] = (dfc.bx*dfc.bx+dfc.by*dfc.by+dfc.bz*dfc.bz)**0.5
dfc
```

```python
df = dfc[(dfc.phi == 0) & (dfc.z > 0)]
df = df.sort_values(["z","rho"])
nx = df["z"].unique().shape[0]
ny = df["rho"].unique().shape[0]
X = df["z"].values.reshape(nx,ny).T
Y = df["rho"].values.reshape(nx,ny).T
Z = df["bmag"].values.reshape(nx,ny).T
fig, ax = plt.subplots(figsize=(12,7))
C = ax.pcolormesh(X,Y,Z,cmap="cividis")
ax.set_aspect(1.0)
fig.colorbar(C,shrink=0.85);
```

```python
# constants 
c = 3.0e8 # m/s

class Coordinate(object):
    def __init__(self,x,y,z):
        self.x = x
        self.y = y
        self.z = z

    def update(self,coord):
        self.x += coord.x
        self.y += coord.y
        self.z += coord.z

    def __repr__(self):
        return "<{:.3f}, {:.3f}, {:.3f}>".format(self.x,self.y,self.z)

    def __mul__(self,factor):
        return Coordinate(
                self.x*factor,
                self.y*factor,
                self.z*factor,
                )

def cross_product(c1,c2):
    x =   c1.y*c2.z - c1.z*c2.y
    y = -(c1.x*c2.z - c1.z*c2.x)
    z =   c1.x*c2.y - c1.y*c2.x
    return Coordinate(x,y,z)

def to_cylindrical_rounded(c1):
    x,y,z = 100.*c1.x,100.*c1.y,100.*c1.z
    rho = np.sqrt(x*x+y*y)
    phi = np.arctan2(y,x)
    rho = int(rho/10)*10
    z = int(z/10)*10
    phi = int(360.*phi/5/(2*3.141592))*5
    return rho,z,phi

class Particle(object):
    def __init__(self,x0,v0,mev):
        self.x0 = x0
        self.v0 = v0
        # self.qoverm = 9.6e7 # 9.6e7 C/kg for proton; 6.4e8 C/kg for pi+
        # self.qoverm = 1.8e11 / (mev/0.5)
        self.qoverm = 8.5e8 / (mev/106.)
        # self.qoverm = 6.4e8 # 9.6e7 C/kg for proton; 6.4e8 C/kg for pi+
        # self.qoverm = 1.76e11 # C/kg for electron
        # self.qoverm = 4.8e7
        # self.qoverm = 1.

    def __repr__(self):
        return "x={}, v={}".format(str(self.x0),str(self.v0))

    def propagate(self,bvec,dt=1.):
        # self.x0.update(1,1,1)
        # self.v0.update(1,1,1)

        dv = cross_product(self.v0,bvec)*dt*self.qoverm
        self.v0.update(dv)
        self.x0.update(self.v0*dt)
```
